```python
# First install required libraries:
# pip install yfinance pandas openpyxl

import yfinance as yf
import pandas as pd

# Enter company ticker
ticker_symbol = "ENB"   # Example: Apple

# Create Yahoo Finance ticker object
company = yf.Ticker(ticker_symbol)

# Download statements
balance_sheet = company.balance_sheet
income_statement = company.financials
cash_flow = company.cashflow

# Check if all are empty
if balance_sheet.empty and income_statement.empty and cash_flow.empty:
    print(f"No financial data found for {ticker_symbol}")
else:
    file_name = f"{ticker_symbol}_financial_statements.xlsx"

    with pd.ExcelWriter(file_name, engine="openpyxl") as writer:
        if not balance_sheet.empty:
            balance_sheet.to_excel(writer, sheet_name="Balance Sheet")
        if not income_statement.empty:
            income_statement.to_excel(writer, sheet_name="Income Statement")
        if not cash_flow.empty:
            cash_flow.to_excel(writer, sheet_name="Cash Flow")

    print(f"Financial statements saved successfully as: {file_name}")
```

```python
from pandas_datareader import data as pdr
import yfinance as yfin
import pandas as pd

# Dates
start_date = '2024-01-01'
end_date = '2026-10-03'

# Tickers
stock_ticker = 'ENB'
market_ticker = '^GSPC'   # S&P 500

# Download data
stock_data = yfin.download(stock_ticker, start=start_date, end=end_date)['Close']
market_data = yfin.download(market_ticker, start=start_date, end=end_date)['Close']

# Save to Excel (multiple sheets)
file_name = "Stock_and_Market_Data.xlsx"

with pd.ExcelWriter(file_name) as writer:
    stock_data.to_excel(writer, sheet_name='ENB Stock Data')
    market_data.to_excel(writer, sheet_name='S&P 500 Data')

print(f"Data saved successfully as: {file_name}")
```
